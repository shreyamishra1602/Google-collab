```python
nitrogen = int(input("Enter your soil nitrogen level= "))
print ( "Soil nitrogen level: ", nitrogen," mg/kg ")
if nitrogen<100 :
  print ("Recommendation - Apply high nitrogen fertilizer\nCosRs.2000/kg\nMessage= Your soil has low nitrogen content and needs supplementation ")
elif 100<= nitrogen<= 250:
  print("Recommendation- Apply medium nitrogen fertilizer\nCost= Rs.1000/kg\nMessage= Your soil has adequate nitrogen but,needs sight supplementation ")
else:
  print("No fertilizer needed, your soil is healthy ")
```

```python
crop= input("Enter the crop name: ")
moisture= int(input(" Enter the percentage of soil moisture :"))
temp= input("Enter the temp(hot/moderate/cool)= ")
print(" Crop= ", crop,"Soil moisture=" , moisture, " Temperature= ",temp)
if crop=="rice "and 50<moisture< 60 and temp== "hot":
  print (" Decision- IRRIGATE NOW\n Reason- Soil moisture is below threshold (60%for hot)\n Estimated water needed= 100mm ")
elif crop== "rice "and moisture<50:
  print (" Decision - IRRIGATE NOW\nReason- Soil moisture is below threshold (50% for moderate/cool)\nEstimated water needed =150 mm")
elif crop== "rice" and moisture >60:
  print (" No need for irrigation ")
elif crop=="wheat"and 30<moisture< 40 and temp== "hot":
  print (" Decision- IRRIGATE NOW\n Reason- Soil moisture is below threshold (40%for hot)\n Estimated water needed= 40mm ")
elif crop== "wheat"and moisture<30:
  print (" Decision - IRRIGATE NOW\nReason- Soil moisture is below threshold (30% for moderate/cool)\nEstimated water needed =50 mm")
elif crop== "wheat" and moisture >40:
  print (" No need for irrigation ")
elif crop=="sugarcane"and 60<moisture< 70 and temp== "hot":
  print (" Decision- IRRIGATE NOW\n Reason- Soil moisture is below threshold (70%for hot)\n Estimated water needed= 70mm ")
elif crop== "sugarcane "and moisture<60:
  print (" Decision - IRRIGATE NOW\nReason- Soil moisture is below threshold (60% for moderate/cool)\n Estimated water needed= 90mm")
elif crop== "sugarcane" and moisture >70:
  print (" No need for irrigation ")
else:
   print("invalid crop")
```

```python
pest= int(input (" Enter the no. of pest per plant: "))
damage= int(input(" Enter the damage percentage : " ))
temp= int(input(" Enter the temp : "))
humidity= int(input("Enter the humidity :" ))
print ("Pests per plant= ",pest, "Damage= ",damage,"Temp = ",temp," Humidity= ",humidity)
if pest>15 or damage>25:
  print( "Severity level-high\n Recommendation -Use strong pesticide as soon as possible\n Cost of intervention=Rs1000/hac")
elif 5< pest <15 or 10<damage<25:
  print ("Severity level-Medium\n Recommendation - Monitor daily\n Consider targeted pesticide spray if severity increases\n Cost of intervention= R,.500/hac ")
else:
  print("Severity level - Low\n Pest damage under control")
```
