```python
import random
from statistics import mode
import numpy as np
#simulate 10 dice rolls#
sample=[random.randint(1,6)for i in range(10)]
print("sample rolls:", sample)
print("sample mean:", sum(sample)/(len(sample)))
print("sample median:",np.median(sample))
print("sample mode:",mode(sample))
print("sample standard deviation:",np.std(sample))
```
