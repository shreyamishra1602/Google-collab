```python
# Soil pH readings (3 fields × 4 samples)
ph_levels = np.array([
 [6.5, 6.8, 6.3, 6.7],
 [7.1, 7.0, 6.9, 7.2],
 [6.2, 6.4, 6.1, 6.3]
])
# Single element: row 1, column 2
print("Single element: row 1, column 2=",ph_levels[1, 2])
# Entire row 0
print("Entire row 0=",ph_levels[0, :])
# Entire column 3
print("Entire column 3=",ph_levels[:, 3])
# First 2 rows, first 3 columns
sub_array = ph_levels[0:2, 0:3]
print("First 2 rows, first 3 columns=",sub_array)
# Every other row, all columns
print("Every other row, all columns=",ph_levels[::2, :])
```

```python
temperatures = np.array([72, 85, 91, 68, 95, 70, 88])
# Create a boolean array (mask)
hot_days = temperatures > 85
print("hot_days=",hot_days)
# Use mask to filter original array
print("hot days=",temperatures[hot_days])
# Can do it in one line
print("hot days=",temperatures[temperatures > 85])
# Combine conditions with & (and), | (or)
moderate = temperatures[(temperatures >= 70) &
(temperatures <= 85)]
print(f"moderate={moderate}")
# Find extreme values (very hot OR very cold)
extreme = temperatures[(temperatures > 90)|(temperatures < 70)]
print(f"extreme={extreme}")
```

```python
# Crop yields in tons
field_a = np.array([120, 135, 142, 128])
field_b = np.array([118, 140, 138, 132])
total_yield = field_a + field_b
print("total_yield =",total_yield)
difference = field_a - field_b
print("difference =",difference)
doubled = field_a * 2
print("doubled=",doubled)

```
