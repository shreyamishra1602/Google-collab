```python
import numpy as np
from collections import Counter
scores=(60,70,46,78,94,48,89,100,35,47,35,46,67,46)
counts = Counter(scores)
# The mode can be one or more values if they have the same highest frequency
mode = [key for key, value in counts.items() if value == max(counts.values())]
print(mode)
```
